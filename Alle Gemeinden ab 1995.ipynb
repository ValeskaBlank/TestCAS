## Funktionen aus allen Schritten
Dann kann ich das für alle anderen Gemeinden wiederholen

```python
#Funktion 1a
def process_data(df1):
    df1["Schätzungsdatum_neu"] = pd.to_datetime(df1["Schätzungsdatum"], format="%d.%m.%Y")
    df1.set_index("Schätzungsdatum_neu", inplace=True)
    df1.rename(columns={"Stand BKI Neuwert\n": "BKI Neuwert"}, inplace=True)
    df1.pop("Schätzungsdatum")
    df1.pop("SIA-Norm")
    df1.rename_axis(None, inplace=True)
    df1.sort_index(inplace=True)
    df1.index = df1.index.to_period("Y")
    df1 = df1.loc["1995":"2020"]
    return df1
```

```python
#Funktion 1b
def combine_and_calculate(df1, dfbki_final):
    df1_merged = df1.combine_first(dfbki_final)
    df1_merged = df1_merged[["Gebäudeart", "Neuwert", "BKI Neuwert", "BKI Jahr", "Kubatur"]]
    df1_merged["Neuwert indexiert"] = df1_merged["BKI Jahr"] * df1_merged["Neuwert"] / df1_merged["BKI Neuwert"]
    df1_merged["Erstellungskosten m3"] = df1_merged["Neuwert indexiert"] / df1_merged["Kubatur"]
    return df1_merged
```

```python
#Funktion 1c
def aggregate(df1_merged):
    df1_auswertung = df1_merged.groupby(df1_merged.index.year)["Erstellungskosten m3"].agg(["min", "max", "median", "count"])
    df1 = df1_auswertung.round(decimals = 1)
    return df1
```

```python
df2 = pd.read_excel("Excel-Files/Eschen.xls")
df2 = process_data(df2)
df2_merged = combine_and_calculate(df2, dfbki_final)
df2_final = aggregate(df2_merged)
df_eschen = df2_final["Gemeinde"] = "Eschen"
df_eschen = df2_final[["Gemeinde", "min", "max", "median", "count"]]
df_eschen.tail(2)
```

Packen wir das wieder in eine Funktion und wiederholen es für alle anderen Gemeinden

```python
#Funktion 1d (für gerade Jahre)
def merge_with_median(df_balzers, df_median):
    df_balzers_inklmedian = df_balzers.merge(df_median, how="inner", left_index=True, right_index=True)
    df_balzers_even_years = df_balzers_inklmedian.index % 2 == 0
    df_balzers_even_years = df_balzers_inklmedian.loc[df_balzers_even_years, :]
    df_balzers_even_years["median_pct_change"] = df_balzers_even_years["median"].pct_change() * 100
    df_balzers_even_years["Medianlohn/CHF_pct_change"] = df_balzers_even_years["Medianlohn/CHF"].pct_change() *100
    return df_balzers_even_years
```
