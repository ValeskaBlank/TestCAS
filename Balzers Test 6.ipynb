```python
import pandas as pd
```

```python
df.index = df.index.to_period("Y")
df.head(2)
```

```python
#Ich will die Jahre 2000 bis 2020 auswerten
df = df.loc["2000":"2020"]
df.head(2)
```

```python
#Neuwert auf das richtige Jahr indexieren. Die Daten sind in einem separaten Excel.
dfbki = pd.read_excel("Excel-Files/BKIJahr.xlsx")
dfbki.head(3)
```

```python
#Ich will das gesamte df anschauen
pd.options.display.max_rows = None
dfbki
```

```python
#Ohje, das geht bis ins Jahr 1939 zurück. Überflüssige Zeilen gemäss Index raushauen
dfbki.drop(dfbki.index[2:63], inplace=True)
dfbki
```

```python
#Indexzeile 0 brauch ich auch nicht. Raus damit
dfbki_temp = dfbki.drop(dfbki.index[0])
dfbki_temp.head(2)
```

```python
#Ok, jetzt kann ich das Gschmäus unten auch noch löschen. Und ich brauch ja nur bis zum Jahr 2020
dfbki_temp = dfbki_temp.iloc[:-6]
dfbki_temp.tail(5)
```

```python
#Looking good! Jetzt putz ich noch ein bisschen weiter. Unnötige Spalten raus 
dfbki_temp1 = dfbki_temp.drop(columns=["Zürcher Index der Wohnbaukosten", "Unnamed: 1"])
dfbki_temp1.tail(2)
```

```python
#Spalten, die ich brauche, sauber anschreiben
dfbki_temp1 = dfbki_temp1.rename(columns={"Unnamed: 2" : "Jahr", "Unnamed: 3" : "BKI Jahr"})
dfbki_temp1.head(2)
```

```python
#neues df mit den Spalten, die ich brauche
dfbki_final = dfbki_temp1[["Jahr", "BKI Jahr"]]
dfbki_final
```

```python
#erste Zeile auch noch raus
dfbki_final.drop(dfbki_final.index[0], inplace=True)
dfbki_final
```

```python
#Welches Format hat eigentlich die Jahresspalte?
dfbki_final.dtypes
```

```python
#Das müssen wir zu datetime machen
dfbki_final["Jahr_neu"] = pd.to_datetime(dfbki_final["Jahr"], format="%Y")
dfbki_final.head(3)
```

```python
#Die alte Jahresspalte kann weg
dfbki_final.pop("Jahr")
dfbki_final.head(2)
```

```python
#Ich will die Jahre wieder als Index
dfbki_final.set_index("Jahr_neu", inplace=True)
dfbki_final.head(2)
```

```python
#Achsentitel kann weg
dfbki_final.rename_axis(None, inplace=True)
dfbki_final.head(3)
```

```python
#Und jetzt wieder nur die Jahre als PeriodIndex, damit ich die df's nachher besser kombinieren kann
dfbki_final.index = dfbki_final.index.to_period("Y")
dfbki_final
```

```python
#Spalte BKI Jahr: Werte will ich als float
dfbki_final["BKI Jahr"] = dfbki_final["BKI Jahr"].astype(float)
dfbki_final.dtypes
```

```python
#Stimmt das Indexformat jetzt?
dfbki_final.index
```

```python
#Jetzt will ich meine beiden df's nach Index mergen. Ich schau nochmal, ob der Index beim ersten df dasselbe Format hat
df.index
```

```python
#Ok, sollte gehen! Let's try
df_merged = df.combine_first(dfbki_final) #diese Methode hat mir Chat GPT verraten. Ist simpler als merge
df_merged.head(5)
```

```python
#Reihenfolge der Spalten anpassen
df_merged = df_merged[["Gemeinde", "Gebäudeart", "Neuwert", "BKI Neuwert", "BKI Jahr", "Kubatur"]]
df_merged.head(3)
```

```python
#neue Werte, die ich brauche, in neuen Spalten berechnen
df_merged["Neuwert indexiert"] = df_merged["BKI Jahr"] * df_merged["Neuwert"] / df_merged["BKI Neuwert"]
df_merged["Erstellungskosten m3"] = df_merged["Neuwert indexiert"] / df_merged["Kubatur"]
df_merged.head(2)
```

```python
#Das hat mir zu viele Nachkommastellen
df_merged["Neuwert indexiert"] = df_merged["Neuwert indexiert"].round(decimals = 1)
df_merged["Erstellungskosten m3"] = df_merged["Erstellungskosten m3"].round(decimals = 1)
df_merged.head(3)
```

```python
#von den Kubikmeterpreisen will ich das Minimum ...
grouped1 = df_merged.groupby(df_merged.index.year)["Erstellungskosten m3"].min()
```

```python
# ... und das Maximum ...
grouped2 = df_merged.groupby(df_merged.index.year)["Erstellungskosten m3"].max()
```

```python
# ... und den Durchschnitt
grouped3 = df_merged.groupby(df_merged.index.year)["Erstellungskosten m3"].mean()
```

```python
#Ergebnisse in ein neues df
df_grouped = pd.concat([grouped1, grouped2, grouped3], axis=1)
df_grouped.head(3)
```
