```python
import pandas as pd
```

```python
df = pd.read_excel("Excel-Files/Balzers.xls")
df["Schätzungsdatum_neu"] = pd.to_datetime(df["Schätzungsdatum"], format="%d.%m.%Y")
df.set_index("Schätzungsdatum_neu", inplace=True)
df.rename(columns={"Stand BKI Neuwert\n": "BKI Neuwert"}, inplace=True)
df.rename_axis(None, inplace=True)
df.pop("Schätzungsdatum")
df.sort_index(inplace=True)
df.head(3)
```

```python
df.index = df.index.to_period("Y")
df.index
```

```python
df2000 = df.loc["2000":"2022"]
df2000.tail(3)
```

```python
dfbki = pd.read_excel("Excel-Files/BKIJahr.xlsx")
dfbki.drop(dfbki.index[2:63], inplace=True)
dfbki2000 = dfbki.drop(dfbki.index[0])
dfbki2000.reset_index(drop=True, inplace=True)
dfbki2000.drop(dfbki2000.index[24], inplace=True)
dfbki_final = dfbki2000.drop(columns=["Zürcher Index der Wohnbaukosten", "Unnamed: 1"])
dfbki_final = dfbki_final.rename(columns={"Unnamed: 2" : "Jahr", "Unnamed: 3" : "BKI Jahr"})
dfbki_final = dfbki_final[["Jahr", "BKI Jahr"]]
dfbki_final.drop(dfbki_final.index[0], inplace=True)
dfbki_final.tail(5)
```

```python
dfbki_final["Jahr_neu"] = pd.to_datetime(dfbki_final["Jahr"], format="%Y")
dfbki_final.head(3)
```

```python
dfbki_final.pop("Jahr")
dfbki_final.set_index("Jahr_neu", inplace=True)
dfbki_final.head(3)
```

```python
dfbki_final.rename_axis(None, inplace=True)
dfbki_final.head(3)
```

```python
dfbki_final.index = dfbki_final.index.to_period("Y")
dfbki_final
```

```python
dfbki_final = dfbki_final.loc[dfbki_final.index.notna()]
dfbki_final.tail(5)
```

```python
dfbki_final["BKI Jahr"] = dfbki_final["BKI Jahr"].astype(float)
dfbki_final.dtypes
```

```python
merged_df = df2000.combine_first(dfbki_final)
```

```python
merged_df = merged_df[["Gebäudeart", "Neuwert", "BKI Neuwert", "BKI Jahr", "Kubatur", "SIA-Norm"]]
merged_df.head(3)
```

```python
merged_df["Neuwert indexiert"] = merged_df["BKI Jahr"] * merged_df["Neuwert"] / merged_df["BKI Neuwert"]
merged_df["Erstellungskosten m3"] = merged_df["Neuwert indexiert"] / merged_df["Kubatur"]
merged_df.head(3)
```
