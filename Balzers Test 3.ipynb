```python
import pandas as pd
```

```python
#file einlesen
df = pd.read_excel("Excel-Files/Balzers.xls")
df.head(2)
```

```python
#Datum in datetime umwandeln
df["Schätzungsdatum_neu"] = pd.to_datetime(df["Schätzungsdatum"], format="%d.%m.%Y")
df.head(2)
```

```python
#Datum als Index setzen
df.set_index("Schätzungsdatum_neu", inplace=True)
df.head(2)
```

```python
#unnötige Spalten löschen und sauber beschriften
df.rename(columns={"Stand BKI Neuwert\n": "BKI Neuwert"}, inplace=True)
df.pop("Schätzungsdatum")
df.pop("SIA-Norm")
df.head(2)
```

```python
#Indexname löschen
df.rename_axis(None, inplace=True)
df.head(2)
```

```python
#Index sortieren
df.sort_index(inplace=True)
df.head(2)
```

```python
#Spalte mit Gemeinde einfügen
df["Gemeinde"] = "Balzers"
df.head(2)
```

```python
#Ich will im Index nur die Jahre
df.index = df.index.to_period("Y")
df.head(2)
```

```python
#Ich will die Jahre 2000 bis 2020 auswerten
df = df.loc["2000":"2020"]
df.head(2)
```

```python
#Neuwert auf das richtige Jahr indexieren. Die Daten sind in einem separaten Excel.
dfbki = pd.read_excel("Excel-Files/BKIJahr.xlsx")
dfbki.head(3)
```

```python
#Ich will das gesamte df anschauen
pd.options.display.max_rows = None
dfbki
```

```python
#Ohje, das geht bis ins Jahr 1939 zurück. Überflüssige Zeilen gemäss Index raushauen
dfbki.drop(dfbki.index[2:63], inplace=True)
dfbki
```

```python
#Indexzeile 0 brauch ich auch nicht. Raus damit
dfbki_temp = dfbki.drop(dfbki.index[0])
dfbki_temp
```

```python
#Ui, ganz unten hats auch noch überflüssige Zeilen. Be gone! Dafür muss ich aber erst den Index neu setzen
dfbki_temp.reset_index(drop=True, inplace=True)
dfbki_temp
```

```python
#Ok, jetzt kann ich das Gschmäus unten auch noch löschen. Und ich brauch ja nur bis zum Jahr 2020
dfbki_temp = dfbki_temp.iloc[:-6]
dfbki_temp.tail(5)
```

```python
#Looking good! Jetzt putz ich noch ein bisschen weiter. Unnötige Spalten raus 
dfbki_temp1 = dfbki_temp.drop(columns=["Zürcher Index der Wohnbaukosten", "Unnamed: 1"])
dfbki_temp1.tail(2)
```

```python
#Spalten, die ich brauche, sauber anschreiben
dfbki_temp1 = dfbki_temp1.rename(columns={"Unnamed: 2" : "Jahr", "Unnamed: 3" : "BKI Jahr"})
dfbki_temp1.head(2)
```

```python
#neues df mit den Spalten, die ich brauche
dfbki_final = dfbki_temp1[["Jahr", "BKI Jahr"]]
dfbki_final
```

```python
#Indexzeile 0 brauch ich auch nicht mehr
dfbki_final.drop(dfbki_final.index[0], inplace=True)
dfbki_final
#Diese Zelle ja nicht mehrmals ausführen, sonst schrumpft alles zusammen!!!
```

```python
#Welches Format hat eigentlich die Jahresspalte?
dfbki_final.dtypes
```

```python
#Das müssen wir zu datetime machen
dfbki_final["Jahr_neu"] = pd.to_datetime(dfbki_final["Jahr"], format="%Y")
dfbki_final.head(3)
```

```python
#Die alte Jahresspalte kann weg
dfbki_final.pop("Jahr")
dfbki_final.head(2)
```

```python
#Ich will die Jahre wieder als Index
dfbki_final.set_index("Jahr_neu", inplace=True)
dfbki_final.head(2)
```

```python
#Achsentitel kann weg
dfbki_final.rename_axis(None, inplace=True)
dfbki_final.head(3)
```

```python
#Und jetzt wieder nur die Jahre als Index, damit ich die df's nachher besser kombinieren kann
dfbki_final.index = dfbki_final.index.to_period("Y")
dfbki_final
```

```python
#Spalte BKI Jahr: Werte will ich als float
dfbki_final["BKI Jahr"] = dfbki_final["BKI Jahr"].astype(float)
dfbki_final.dtypes
```

```python
#Und was ist der Index?
dfbki_final.index
```

```python
#Jetzt will ich meine beiden df's nach Index mergen. Ich schau nochmal, ob der Index beim ersten df dasselbe Format hat
df.index
```

```python
#Ok, sollte gehen! Let's try
df_merged = df.combine_first(dfbki_final) #diese Methode hat mir Chat GPT verraten. Ist simpler als merge
df_merged.head(5)
```

```python
#Reihenfolge der Spalten anpassen
df_merged = df_merged[["Gebäudeart", "Neuwert", "BKI Neuwert", "BKI Jahr", "Kubatur"]]
df_merged.head(3)
```

```python
#neue Werte, die ich brauche, in neuen Spalten berechnen
df_merged["Neuwert indexiert"] = df_merged["BKI Jahr"] * df_merged["Neuwert"] / df_merged["BKI Neuwert"]
df_merged["Erstellungskosten m3"] = df_merged["Neuwert indexiert"] / df_merged["Kubatur"]
df_merged.head(2)
```

```python
#Das hat mir zu viele Nachkommastellen
df_merged["Neuwert indexiert"] = df_merged["Neuwert indexiert"].round(decimals = 1)
df_merged["Erstellungskosten m3"] = df_merged["Erstellungskosten m3"].round(decimals = 1)
df_merged.head(3)
```

```python
#von den Kubikmeterpreisen will ich das Minimum ...
df_merged.groupby(df_merged.index.year)["Erstellungskosten m3"].min()
```

```python
# ... und das Maximum
df_merged.groupby(df_merged.index.year)["Erstellungskosten m3"].max()
```

```python
#neue Spalte mit den Ergebnissen aus dem groupby (min/max)
df_merged["min. Erstellungskosten m3 Jahr"] = df_merged.groupby(df_merged.index.year)["Erstellungskosten m3"].transform("min").round(decimals = 1)
df_merged["max. Erstellungskosten m3 Jahr"] = df_merged.groupby(df_merged.index.year)["Erstellungskosten m3"].transform("max").round(decimals = 1)
df_merged
```

```python
#wie viele Werte pro Jahr hat es eigentlich im df? Das ist die Anzahl hinterlegter Gebäude
df_merged.index.value_counts()
```

```python
#Packen wir das auch noch in eine neue Spalte
df_merged["Anz. hinterlegte Gebäude/Jahr"] = df_merged.index.value_counts()
df_merged.head(2)
```

```python
#df ist schon wieder viel zu gross! Machen wir es schlanker
df_final = df_merged[["min. Erstellungskosten m3 Jahr", "max. Erstellungskosten m3 Jahr", "Anz. hinterlegte Gebäude/Jahr"]]
df_final
```

```python
df_final.index
#huh, warum hat's das Format des Index geändert? So möchte ich das nicht
```

```python
#Ändern wir das in Datetime ...
df_final.index = pd.to_datetime(df_final.index, format="%Y")
df_final.head(2)
```

```python
#... und dann in PeriodIndex, damit alles wieder einheitlich ist
df_final.index = df_final.index.to_period("Y")
df_final

#diese Zeile nie zweimal ausführen, verhaut wieder alles
```
